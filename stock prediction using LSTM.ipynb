```python
import yfinance as yf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout

# Suppress TensorFlow warnings
import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3'
```

```python
def get_stock_data(ticker, start="2015-01-01", end="2024-01-01"):
    stock = yf.Ticker(ticker)
    df = stock.history(start=start, end=end)

    if df.empty:
        print(f"No data available for {ticker}")
        return None

    return df[['Close']]  # Using only closing prices
```

```python
def preprocess_data(df, time_steps=60):
    scaler = MinMaxScaler(feature_range=(0, 1))
    scaled_data = scaler.fit_transform(df)

    X, y = [], []
    for i in range(time_steps, len(scaled_data)):
        X.append(scaled_data[i - time_steps:i, 0])
        y.append(scaled_data[i, 0])

    X, y = np.array(X), np.array(y)
    X = np.reshape(X, (X.shape[0], X.shape[1], 1))  # Reshape for LSTM

    return X, y, scaler
```

```python
def build_lstm_model(input_shape):
    model = Sequential([
        LSTM(units=50, return_sequences=True, input_shape=input_shape),
        Dropout(0.2),
        LSTM(units=50, return_sequences=False),
        Dropout(0.2),
        Dense(units=25),
        Dense(units=1)
    ])

    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

```python
def train_model(model, X_train, y_train, epochs=20, batch_size=32):
    history = model.fit(X_train, y_train, epochs=epochs, batch_size=batch_size, validation_split=0.1, verbose=1)
    return history
```

```python
def make_predictions(model, X_test, scaler):
    predictions = model.predict(X_test)
    return scaler.inverse_transform(predictions)
```

```python
def plot_predictions(df, y_test, predictions):
    plt.figure(figsize=(12, 6))
    plt.plot(df.index[-len(y_test):], y_test, color='blue', label="Actual Prices")
    plt.plot(df.index[-len(y_test):], predictions, color='red', linestyle='dashed', label="Predicted Prices")
    plt.xlabel("Date")
    plt.ylabel("Stock Price")
    plt.legend()
    plt.show()
```

```python
if __name__ == "__main__":
    ticker = input("Enter stock ticker symbol: ")
    df = get_stock_data(ticker)

    if df is not None:
        train_size = int(len(df) * 0.8)
        train_data, test_data = df[:train_size], df[train_size:]

        X_train, y_train, scaler = preprocess_data(train_data)
        X_test, y_test, _ = preprocess_data(test_data)

        model = build_lstm_model((X_train.shape[1], 1))
        train_model(model, X_train, y_train)

        predictions = make_predictions(model, X_test, scaler)

        # Inverse transform y_test to original scale
        y_test = scaler.inverse_transform(y_test.reshape(-1, 1))

        # Evaluate Model
        evaluate_model(y_test, predictions)

        # Plot results
        plot_predictions(df, y_test, predictions)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error

def evaluate_model(y_test, predictions):
    mse = mean_squared_error(y_test, predictions)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_test, predictions)

    print(f"Mean Squared Error (MSE): {mse:.4f}")
    print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
    print(f"Mean Absolute Error (MAE): {mae:.4f}")

    return mse, rmse, mae
```

```python
import numpy as np
import pandas as pd
import datetime
import yfinance as yf
from sklearn.preprocessing import StandardScaler

def fetch_yfinance_data(ticker, start_date, end_date):
    """
    Fetch historical stock data from Yahoo Finance.
    - Uses '1d' interval (since '1h' data is unavailable beyond 60 days).
    - Sets Date-Time as a single index.
    """
    df = yf.download(ticker, start=start_date, end=end_date, interval="1d", auto_adjust=True)
    df.index = df.index.tz_localize(None)  # Remove timezone info
    return df

def df_to_windowed_df(dataframe, n=3):
    """
    Converts a time series dataframe into a windowed format.
    - Uses `n` past values to predict the next one.
    - Sets Date-Time as a single index.
    """
    dates = []
    X, Y = [], []

    for i in range(n, len(dataframe)):
        df_subset = dataframe.iloc[i-n:i+1]

        values = df_subset['Close'].to_numpy()
        x, y = values[:-1], values[-1]

        dates.append(dataframe.index[i])
        X.append(x)
        Y.append(y)

    # Convert X to a proper NumPy array
    X = np.array(X).reshape(-1, n)  # Ensure 2D shape

    ret_df = pd.DataFrame(index=dates)

    for i in range(n):
        ret_df[f'Target-{n-i}'] = X[:, i]

    ret_df['Target'] = Y

    return ret_df

# Fetch stock data from Yahoo Finance
df = fetch_yfinance_data("AAPL", "2024-01-01", "2024-03-15")  # Adjust date range

# Convert to windowed format with Date-Time as index
windowed_df = df_to_windowed_df(df, n=3)

# Normalize the features
scaler = StandardScaler()
X_columns = [col for col in windowed_df.columns if "Target-" in col]
windowed_df[X_columns] = scaler.fit_transform(windowed_df[X_columns])

# Split into training and testing sets (80% train, 20% test)
split = int(len(windowed_df) * 0.8)
train_df = windowed_df.iloc[:split]
test_df = windowed_df.iloc[split:]

# Extract final X and y
X_train, y_train = train_df[X_columns].values, train_df["Target"].values
X_test, y_test = test_df[X_columns].values, test_df["Target"].values

# Print results
print("Train set size:", X_train.shape, y_train.shape)
print("Test set size:", X_test.shape, y_test.shape)
print(windowed_df.head())
```

# Taking a year's data and prediciting the price

```python
import numpy as np
import pandas as pd
import datetime
import yfinance as yf
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam
from tensorflow.keras import layers

# ✅ Download Stock Data
ticker = "TSLA"
df = yf.download(ticker, start="2023-01-01", end="2023-12-31", interval="1d")  # Use daily data taking a smaller dataset a year's data

# ✅ Convert Date to Datetime Index
df.index = pd.to_datetime(df.index)

# ✅ Function to Convert Data into Windowed Format
def df_to_windowed_df(dataframe, n=3):
    dates = []
    X, Y = [], []

    for i in range(len(dataframe) - n):
        x = dataframe['Close'].iloc[i:i+n].values
        y = dataframe['Close'].iloc[i+n]

        dates.append(dataframe.index[i + n])
        X.append(x)
        Y.append(y)

    ret_df = pd.DataFrame({'Target Date': dates})
    X = np.array(X)

    for i in range(n):
        ret_df[f'Target-{n-i}'] = X[:, i]

    ret_df['Target'] = Y
    return ret_df

# ✅ Convert to Windowed Data
windowed_df = df_to_windowed_df(df, n=3)

# ✅ Extract X (features) & y (target)
X = windowed_df.iloc[:, 1:-1].values  # All columns except 'Target Date' & 'Target'
y = windowed_df['Target'].values
dates = windowed_df['Target Date'].values  # Extract dates

# ✅ Split Data (80% Train, 10% Validation, 10% Test)
q_80 = int(len(dates) * 0.8)
q_90 = int(len(dates) * 0.9)

dates_train, X_train, y_train = dates[:q_80], X[:q_80], y[:q_80]
dates_val, X_val, y_val = dates[q_80:q_90], X[q_80:q_90], y[q_80:q_90]
dates_test, X_test, y_test = dates[q_90:], X[q_90:], y[q_90:]

# ✅ Reshape for LSTM (samples, timesteps, features)
X_train = np.array(X_train, dtype=np.float32).reshape((X_train.shape[0], X_train.shape[1], 1))
X_val = np.array(X_val, dtype=np.float32).reshape((X_val.shape[0], X_val.shape[1], 1))
X_test = np.array(X_test, dtype=np.float32).reshape((X_test.shape[0], X_test.shape[1], 1))

y_train = np.array(y_train, dtype=np.float32)
y_val = np.array(y_val, dtype=np.float32)
y_test = np.array(y_test, dtype=np.float32)

# ✅ Define LSTM Model
model = Sequential([
    layers.Input((3, 1)),
    layers.LSTM(64),
    layers.Dense(32, activation='relu'),
    layers.Dense(32, activation='relu'),
    layers.Dense(1)
])

model.compile(loss='mse',
              optimizer=Adam(learning_rate=0.001),
              metrics=['mean_absolute_error'])

# ✅ Train Model
history = model.fit(X_train, y_train, validation_data=(X_val, y_val), epochs=500)

# ✅ Training Predictions
train_predictions = model.predict(X_train).flatten()

plt.figure(figsize=(10, 5))
plt.plot(dates_train, train_predictions, label="Training Predictions")
plt.plot(dates_train, y_train, label="Training Observations")
plt.legend()
plt.xlabel("Date")
plt.ylabel("Stock Price")
plt.title("LSTM Model Predictions on Training Data")
plt.xticks(rotation=45)
plt.show()
```

```python
val_predictions = model.predict(X_val).flatten()

plt.plot(dates_val, val_predictions)
plt.plot(dates_val, y_val)
plt.legend(['Validation Predictions', 'Validation Observations'])
```

```python
test_predictions = model.predict(X_test).flatten()

plt.plot(dates_test, test_predictions)
plt.plot(dates_test, y_test)
plt.legend(['Testing Predictions', 'Testing Observations'])
```
